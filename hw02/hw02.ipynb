## Задание 1

Даны два вектора в трехмерном пространстве: (20,20,20) и (0,0,-20)
1. Найдите их сумму. (на листочке)
2. Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

#### 1.1 

[20, 20, 20,] + [0, 0, -20] = [20, 20, 0]

#### 1.2

```python
def compVSize(x: np.array, y: np.array):
    tmp = 0
    for i in range(x.size):
        tmp += (y[i] - x[i])**2
    return math.sqrt(tmp)
```

```python
x = np.array([5, 0, 0])
y = np.array([0, 5, 0])
```

```python
s = compVSize(x, y)
```

## Задание 2

Почему прямые не кажутся перпендикулярными?

```python
x = np.linspace(-5, 5, 21)
y1 = 3*x+1
y2 = (-1/3)*x+1
```

```python
#rcParams['figure.figsize'] = 5, 5
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### Ответ: из-за разного масштаба осей

```python
plt.gca().set_aspect('equal', adjustable='box')
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
