## Задание 1

Даны два вектора в трехмерном пространстве: (20,20,20) и (0,0,-20)
1. Найдите их сумму. (на листочке)
2. Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

#### 1.1 

[20, 20, 20,] + [0, 0, -20] = [20, 20, 0]

#### 1.2

```python
def compVSize(x: np.array, y: np.array):
    tmp = 0
    for i in range(x.size):
        tmp += (y[i] - x[i])**2
    return math.sqrt(tmp)
```

```python
x = np.array([5, 0, 0])
y = np.array([0, 5, 0])
```

```python
s = compVSize(x, y)
```

### 3.1

`(x – a)**2 + (y – b)**2 = R**2`
тогда
`y = b [+/-] SQRT(R**2 - (x – a)**2)`

```python
a = 5
b = 5
r = 10
x = np.linspace( a - r, a + r, 1000)
y1 = []
y2 = []
p = 0.25
for i in range(1000):
    y1.append(b + math.sqrt(abs(r**2 - (x[i] - a)**2)))
    y2.append(b - math.sqrt(abs(r**2 - (x[i] - a)**2)))
```

```python
plt.gca().set_aspect('equal', adjustable='box')
plt.plot(x, y1)
plt.plot(x, y2)
plt.grid(True)
plt.show()
```

`x**2/a**2 + y**2/b**2 = 1`, тогда `y = SQRT((1 - x**2/a**2)b**2)`

```python
a = 10
b = 5
x = np.linspace( -a, a, 1000)
y1 = []
y2 = []
p = 0.25
for i in range(1000):
    y1.append(math.sqrt(abs(1 - x[i]**2/a**2)*b**2))
    y2.append(-math.sqrt(abs(1 - x[i]**2/a**2)*b**2))
```
